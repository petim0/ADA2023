**2.5—** Calculate the mean and the 95% confidence interval for the salary of left-handed and right-handed pitchers aggregating data across all years. Print the estimates. Additionally, present the estimates and the confidence intervals in a [point plot](https://seaborn.pydata.org/_images/pointplot_7_0.png). Calculate your confidence intervals using the standard error of the mean and ignoring missing values.

```python
grouped_salary = df[["yearID", "throws", "salary"]].groupby(["yearID", "throws"])
grouped_salary.head()
```

```python
grouped_salary_mean = grouped_salary.mean().add_suffix('_mean')
grouped_salary_sem = grouped_salary.sem().add_suffix('_sem')

grouped_salary = pd.concat([grouped_salary_mean, grouped_salary_sem], axis=1)

grouped_salary["ci"] = 1.96 * grouped_salary["salary_sem"]
grouped_salary["upp_ci"] = grouped_salary["salary_mean"] + 1.96 * grouped_salary["salary_sem"]
grouped_salary["low_ci"] = grouped_salary["salary_mean"] - 1.96 * grouped_salary["salary_sem"]


grouped_salary.head(100)
```

```python
sns.set(rc={'figure.figsize':(14,9)})
sns.pointplot(data=df, x="yearID", y="salary", hue="throws", errorbar=("ci", 95), dodge=0.1, capsize=0.1)
```
